Task 2:Find the 98th percentile of BMI of your sample and the population and compare the results using charts. 

```python


# Take a random sample of 25 observations
sample_data = df.sample(n=25)

# Calculate the 98th percentile of BMI for the sample and population
sample_percentile = np.percentile(sample_data["BMI"], 98)
pop_percentile = np.percentile(df["BMI"], 98)
print("Percentile value of the sample :" ,sample_percentile);
print("Percentile value of the population :" ,pop_percentile)


```

```python
# Create a bar chart to compare the sample and population percentiles
plt.bar(["Sample 98th Percentile", "Population 98th Percentile"], [sample_percentile, pop_percentile])
plt.title("Comparison of Sample and Population 98th Percentiles")
plt.ylabel("BMI")
plt.show()
```

Compare the BMI values in details

Using pie charts:
 To show data that are above and below the percentages.

```python


# Create a pie chart to compare the proportion of individuals above the 98th percentile of BMI in the sample and population data
labels = ["Below 98th percentile", "Above 98th percentile"]
colors = ["green", "red"]
explode = [0, 0.1]
sample_above_98th = np.mean(sample_data["BMI"] > sample_percentile)
pop_above_98th = np.mean(df["BMI"] > pop_percentile)
plt.pie([1 - sample_above_98th, sample_above_98th], labels=labels, colors=colors, explode=explode, autopct='%1.1f%%')
plt.title("Proportion Above 98th Percentile of BMI in Sample Data")
plt.show()

plt.pie([1 - pop_above_98th, pop_above_98th], labels=labels, colors=colors, explode=explode, autopct='%1.1f%%')
plt.title("Proportion Above 98th Percentile of BMI in Population Data")
plt.show()
```

Task3:

Using bootstrap (replace= True), create 500 samples (of 150 observation each) from the population and find the average mean, standard deviation and percentile for BloodPressure and compare this with these statistics from the population for the same variable. Again, you should create charts for this comparison. Report on your findings

```python
# Define a function to generate a bootstrap sample of size n from a given data set
def bootstrap_sample(data, n):
    return data.sample(n=n, replace=True)
```

```python
# Set the number of bootstrap samples to generate
num_samples = 500
```

```python
# Set the sample size for each bootstrap sample
sample_size = 150
```

```python
# Generate the bootstrap samples and calculate summary statistics for BloodPressure in each sample
bootstrap_means = []
bootstrap_stds = []
bootstrap_percentiles = []
for i in range(num_samples):
    sample = bootstrap_sample(df, sample_size)
    bootstrap_means.append(sample["BloodPressure"].mean())
    bootstrap_stds.append(sample["BloodPressure"].std())
    bootstrap_percentiles.append(np.percentile(sample["BloodPressure"], 95))
    
```

```python
# Calculate summary statistics for BloodPressure in the population data
pop_mean = df["BloodPressure"].mean()
pop_std = df["BloodPressure"].std()
pop_percentile = np.percentile(df["BloodPressure"], 98)
print("Population Mean :",pop_mean)
print("Population Standard Deviation :",pop_std)
print("Population Percentile :",pop_percentile)
```

```python
# Create a bar graph to compare the mean BloodPressure values in the bootstrap samples with the population mean
plt.bar(["Population Mean", "Bootstrap Means"], [pop_mean, np.mean(bootstrap_means)])
plt.xlabel("Statistic")
plt.ylabel("BloodPressure")
plt.title("Comparison of Mean BloodPressure")
plt.show()
```

```python
# Create a bar graph to compare the standard deviation values in the bootstrap samples with the population standard deviation
plt.bar(["Population Standard Deviation", "Bootstrap Standard Deviations"], [pop_std, np.mean(bootstrap_stds)])
plt.xlabel("Statistic")
plt.ylabel("BloodPressure")
plt.title("Comparison of Standard Deviation of BloodPressure")
plt.show()
```

```python
# Create a bar graph to compare the 95th percentile BloodPressure values in the bootstrap samples with the population 95th percentile
plt.bar(["Population 95th Percentile", "Bootstrap 95th Percentiles"], [pop_percentile, np.mean(bootstrap_percentiles)])
plt.xlabel("Statistic")
plt.ylabel("BloodPressure")
plt.title("Comparison of 95th Percentile BloodPressure")
plt.show()
```

```python
# Create a list of labels for the pie chart
labels = ["Mean Population", "Mean Bootstrap Samples"]

# Create a list of values for the pie chart
values = [pop_mean, np.mean(bootstrap_means)]

# Create the pie chart
plt.pie(values, labels=labels, autopct="%1.1f%%")

# Add a title to the pie chart
plt.title("Comparison of Mean BloodPressure")

# Display the pie chart
plt.show()
```

```python
# Create a list of labels for the pie chart
labels = ["Population Standard Deviation", "Bootstrap Standard Deviation"]

# Create a list of values for the pie chart
values = [pop_std, np.mean(bootstrap_stds)]

# Create the pie chart
plt.pie(values, labels=labels, autopct="%1.1f%%")

# Add a title to the pie chart
plt.title("Comparison of Standard Deviation BloodPressure")

# Display the pie chart
plt.show()
```

```python
# Create a list of labels for the pie chart
labels = ["Population", "Bootstrap Samples"]

# Create a list of values for the pie chart
values = [pop_percentile, np.mean(bootstrap_percentiles)]

# Create the pie chart
plt.pie(values, labels=labels, autopct="%1.1f%%")

# Add a title to the pie chart
plt.title("Comparison of  Percentile Blood Pressure")

# Display the pie chart
plt.show()
```
